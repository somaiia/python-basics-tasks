### Question 1: Count Vowels in a String

Write a function `count_vowels` that takes a string and returns the count of vowels (a, e, i, o, u).
The function should be case-insensitive.

```python
# Question1: count vowels in a string
def count_vowels(text):
    count = 0
    vowels = "aeiouAEIOU"
    for char in text:
        if char in vowels:
            count += 1
    return count

text = "Hello Word"
print("Number of vowels in the text:", count_vowels(text))
```

### Question 2: Find the Maximum Value in a List

Write a function `find_maximum` that takes a list of integers and returns the maximum value.
Do **not** use Python's built-in `max()`. Return `None` for empty lists.

```python
# Question2: the maximum value in a list
def find_maximum(numbers):
    if not numbers:
        return None
    max_value = numbers[0]
    for num in numbers:
        if num > max_value:
            max_value = num
    return max_value

empty_list = []
print("maximum value in the list:", find_maximum(empty_list))

Numbers = [1, 10, 15, 8, 18]
print("maximum value in the list:", find_maximum(Numbers))
```
