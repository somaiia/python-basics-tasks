### Q1: Count Vowels in a String

Write a function `count_vowels` that takes a string and returns the count of vowels (a, e, i, o, u).
The function should be case-insensitive.

```python
# Example Input:
# text = "Hello World"
# Expected Output: 3

def count_vowels(text):
    vowels = "aeiouAEIOU"
    return sum(1 for ch in text if ch in vowels)

# Test
text = "Hello World"
print(count_vowels(text))  # 3
```

### Q2: Find the Maximum Value in a List

Write a function `find_max` that takes a list of integers and returns the maximum value.
Do **not** use Python's built-in `max()`. Return `None` for empty lists.

```python
# Example Input:
# numbers = [3, 5, 2, 8, 1]
# Expected Output: 8

def find_max(numbers):
    if not numbers:
        return None
    maximum = numbers[0]
    for num in numbers[1:]:
        if num > maximum:
            maximum = num
    return maximum

# Test
numbers = [3, 5, 2, 8, 1]
print(find_max(numbers))  # 8
```
